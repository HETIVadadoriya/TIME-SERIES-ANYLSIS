```python
def flight_dep_time(x):
    if(x>4) and (x<=8):
        return "Early morning"
        
    elif(x>8) and (x<=12):
        return "Morning"
        
    elif (x>12) and (x<=16):
        return "Noon"
        
    elif (x>16) and (x<=20):
        return "Evening"

    elif (x>20) and (x<=24):
        return "Night"
    else:
        return "Late night"
```

```python
def preprocess_duration(x):
    if 'h' not in x:
       x = '0h' + ' ' + x
    elif 'm' not in x:
        x = x + ' ' + '0m'

    return x
```
